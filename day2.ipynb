```python
students_marks = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "David": 65,
    "Eve": 95
}

# Display the dictionary
print(students_marks)
```

```python
marks = students_marks.values()
average_marks = sum(marks) / len(marks)

print(f"The average marks for the students is: {average_marks:.2f}")
```

```python
scores = students_marks.values()
highest_score = max(scores)
lowest_score = min(scores)

print(f"The highest score is: {highest_score}")
print(f"The lowest score is: {lowest_score}")
```

```python
print("Students scoring above the average:")
for student, marks in students_marks.items():
    if marks > average_marks:
        print(f"- {student}: {marks}")
```

```python
student_name_search = input("Enter the student's name to search: ")

if student_name_search in students_marks:
    print(f"{student_name_search}'s marks: {students_marks[student_name_search]}")
else:
    print(f"Student '{student_name_search}' not found in the list.")
```

```python
student_to_update = input("Enter the name of the student whose mark you want to update: ")
if student_to_update in students_marks:
    try:
        new_mark = float(input(f"Enter the new mark for {student_to_update}: "))
        students_marks[student_to_update] = new_mark
        print(f"Updated {student_to_update}'s mark to {new_mark}.")
        print("Revised student marks:", students_marks)
    except ValueError:
        print("Invalid mark entered. Please enter a number.")
else:
    print(f"Student '{student_to_update}' not found.")
```

```python
# Recalculate the average marks after update
marks_after_update = students_marks.values()
revised_average_marks = sum(marks_after_update) / len(marks_after_update)

print(f"The revised average marks for the students is: {revised_average_marks:.2f}")
```
